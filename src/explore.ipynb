```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import f_classif, SelectKBest
import warnings
warnings.filterwarnings("ignore")
import numpy as np
from scipy import stats
from scipy.stats import lognorm
```

```python
train_data = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/machine-learning-content/master/assets/clean_titanic_train.csv")
test_data = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/machine-learning-content/master/assets/clean_titanic_test.csv")

train_data.head()
```

```python
num_variables = ["Pclass", "Fare", "Sex_n", "Embarked_n", "FamMembers"]

X_train = train_data.drop("Survived", axis = 1)[num_variables]
y_train = train_data["Survived"]
```

```python
X_test = test_data.drop("Survived", axis = 1)[num_variables]
y_test = test_data["Survived"]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X_train)

X_train_scal = scaler.transform(X_train)
X_train_scal = pd.DataFrame(X_train_scal, index = X_train.index, columns = num_variables)

X_test_scal = scaler.transform(X_test)
X_test_scal = pd.DataFrame(X_test_scal, index = X_test.index, columns = num_variables)

X_train_scal.head()
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train_scal, y_train)
```

```python
y_pred = model.predict(X_test_scal)
y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```
